```python
# Crear dataframe
import pandas as pd

ruta_archivo = '/content/drive/MyDrive/Colab Data/applicant_list_7.txt'

nombres_columnas = ['nombre', 'apellido', 'EPH', 'ECH', 'EMA', 'ECS', 'ESP', 'OP1', 'OP2', 'OP3']

df = pd.read_csv(ruta_archivo, sep=' ', names=nombres_columnas)

df
```

```python
# Mostrar tamaño y descripción de dataframe
print(df.shape)
df.info()
```

```python
# Mostrar opciones de carreras por aspirante
display(df[['OP1', 'OP2', 'OP3']])
```

```python
# Unir columna de "nombre" y "apellido"
df['nombre_completo'] = df['nombre'] + ' ' + df['apellido']
df

```

```python
# Función para calcular puntaje de admisión
def calcular_puntaje_admision(linea, opcion):
  departamento = linea[opcion]
  if departamento == 'Physics':
    puntaje = (linea['EPH'] + linea['EMA'] + linea['ESP']) / 3
  elif departamento == 'Mathematics':
    puntaje = (linea['EMA']) + linea['ECS'] + linea['ESP'] / 3
  elif departamento == 'Engineering':
    puntaje = (linea['EMA'] + linea['ECS'] + linea['EPH'] + linea['ESP']) / 4
  elif departamento == 'Chemistry':
    puntaje = (linea['ECH'] + linea['EPH'] + linea['ESP']) / 3
  elif departamento == 'Biotech':
    puntaje = (linea['EMA'] + linea['ECS'] + linea['ECH'] + linea['ESP']) / 4
  else:
    return None
  return round(puntaje, 2)

df['OP1_score'] = df.apply(lambda x: calcular_puntaje_admision(x, 'OP1'), axis=1)
df['OP1_score'] = df.apply(lambda x: calcular_puntaje_admision(x, 'OP2'), axis=1)
df['OP1_score'] = df.apply(lambda x: calcular_puntaje_admision(x, 'OP3'), axis=1)
df
```

```python
# Cupo máximo de admitidos por departamento seleccionado
N = int(input("Ingrese el cupo maximo por departamento: "))
departamento_seleccionado = input("Ingrese el departamento: ")
```

```python
# Identificar aspirantes que concuerden con el departamento seleccionado
aspirantes_departamento = df[(df['OP1'] == departamento_seleccionado)]

# Ordenar de forma descendente aspirantes por promedio
aspirantes_ordenados = aspirantes_departamento.sort_values(by='OP1_score', ascending=False)

# Crear tabla de los 'N' aspirantes admitidos
tabla_aspirantes_admitidos = aspirantes_ordenados.head(N)

tabla_en_formato = pd.DataFrame(
    {
      'nombre_completo': tabla_aspirantes_admitidos['nombre_completo'],
      'departamento': tabla_aspirantes_admitidos['OP1'],
      'puntaje': tabla_aspirantes_admitidos['OP1_score'],
      'opcion': 'OP1'
    }
)

display(tabla_en_formato)
print(tabla_en_formato.shape)
```
